```python
#!/user/bin/env python
# coding:utf-8

#Parse items into ceo_list

import selenium
import time
from selenium import webdriver
import glob
files = glob.glob('/home/guan/Data/ceo_education/ceo_pages/1_10000/*.html')

phantomjs_path = '/home/guan/Software/phantomjs-2.1.1-linux-x86_64/bin/phantomjs'
outfile = open('/home/guan/Data/ceo_education/ceo_education_sample.csv', 'w')
ds = webdriver.PhantomJS(executable_path=phantomjs_path)
print >> outfile, 'name|title|company|education'
for n, file_name in enumerate(files):
    if n % 10 == 0:
        print n
    ds.get(file_name)
    education = ''
    try:
        education = ds.find_element_by_xpath('//div[@class="education first last section"]')
    except:
        continue
    education_results = education.find_elements_by_xpath('.//ul[@class="records"]/li')
    edu_str = ''
    for item in education_results:
        edu_str = edu_str + ';' + item.text
    edu_str = edu_str.strip(';')
    header = ds.find_element_by_xpath('//div[@class="header"]')
    try:
        ceo_name = header.find_element_by_xpath('.//h1[@class="name"]').text
    except:
        continue
    ceo_title = ''
    try:
        ceo_title = header.find_element_by_xpath('.//div/h2').text.strip(',')
    except:
        pass
    ceo_company = ''
    try:
        ceo_company = header.find_element_by_xpath('.//div/a').text
    except:
        pass
    
    strtmp = ceo_name + '|' + ceo_title + '|'  + ceo_company + '|' + edu_str
    print strtmp
    print >> outfile, strtmp.encode('utf-8')
```

```python
from bs4 import BeautifulSoup
import glob
import json
import collections
files = glob.glob('/home/guan/Data/ceo_education/ceo_pages/test/*.html')
outfile = open('/home/guan/Data/ceo_education/ceo_pages/ceo_dicts/ceo_dicts', 'w')

for html_file in files:
    print html_file
    ceo_dic = collections.OrderedDict()
    html = open(html_file)
    soup = BeautifulSoup(html,'html.parser',from_encoding='utf-8')
    name = soup.find('h1', {'class': 'name'}).text.strip()
    #print name.text
    if name == '':
        continue
    ceo_dic[u'name'] = name
    title = soup.find('h2', {'class': 'title'})
    #print title.text
    ceo_dic[u'title'] = title.text.strip(',')
    company = title.find_next('a').text
    #print company
    ceo_dic[u'company'] = company.strip()
    
    
    records = soup.find_all('h2', {'class': 'module_title'})
    for record in records:
        #ceo_dic[record.text.strip()] = []
        #print record.text.strip()
        #print '================='
        items = record.find_next_siblings('div')
        item_content = []
        for item in items:
            if item.text.strip() != '':
                item_content.append(item.text.strip())
            
            #print '+++++++++++++++'
        if item_content != []:
            ceo_dic[record.text.strip()] = item_content
    html.close()
    #print ceo_dic
    print >> outfile, json.dumps(ceo_dic.items()) 
    #print >> outfile, json_format


outfile.close()
        
    
    
```
