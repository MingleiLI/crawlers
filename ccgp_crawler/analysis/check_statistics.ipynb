```python
#coding = utf-8
infile = open('ccgp_it_result.csv')
supplier_dic = {}
consumer_dic = {}
amount_list = []
year_dic = {}
provinces = ['北京', '天津市', '上海', '重庆', '河北', '山西', '辽宁', '吉林', '黑龙江', '江苏', '浙江', '安徽', '福建', '江西', '山东', '河南', '湖北', '湖南', '广东', '海南', '四川', '贵州', '云南', '陕西','甘肃','青海','台湾','内蒙','广西','西藏','宁夏','新疆']
institute = ['局','中心', '小学','中学','大学','学校','学院','公司','处', '法院','处','医院','银行','所', '海关']
prov_dic = {}
#outfile_amount = open('amount', 'w')
def change_amount(input_str):
    strtmp = ''
    for loc in input_str:
        if loc.isdigit() or loc == '.':
            strtmp = strtmp + loc
    try:
        strtmp = float(strtmp)
        if strtmp < 1000:
            strtmp = strtmp * 10000
        return strtmp
    except:
        return 0

def get_year(input_str):
    input_str = input_str.strip().split('年')
    try:
        return int(input_str[0].strip())
    except:
        return 0

def get_province(input_str):
    for item in provinces:
        if item in input_str:
            return item
    return 0

def get_consumer(input_str):
    for item in institute:
        if item in input_str:
            input_str = input_str.strip().split(item)
            return input_str[0].strip() + item       
    return 'NA'
    

for line in infile:
    title, consumer, time, supplier, amount, content  = line.strip().split('\t')
    
    if time != 'NA':
        time = get_year(time)
        if time != 0:
            if time not in year_dic:
                year_dic[time] = 1
            else:
                year_dic[time] += 1
    if amount != 'NA':
        amount = change_amount(amount)
        
        #print >> outfile_amount, amount
        if amount > 10000 and amount < 10000000:
            amount_list.append(amount)
        else:
            amount = 0
    else:
        continue
    if title != 'NA':
        prov = get_province(title)
        if prov != 0:
            if prov not in prov_dic:
                prov_dic[prov] = amount
            else:
                prov_dic[prov] += amount
    if supplier != 'NA':
        if supplier not in supplier_dic:
            #supplier_dic[supplier] = amount
            supplier_dic[supplier] = 1
        else:
            #supplier_dic[supplier] += amount
            supplier_dic[supplier] += 1

    if consumer == 'NA':
        consumer = get_consumer(title)
        
    if consumer != 'NA':
        if consumer not in consumer_dic:
            consumer_dic[consumer] = amount
            #consumer_dic[consumer] = 1
        else:
            consumer_dic[consumer] += amount
            #consumer_dic[consumer] += 1
    
        
            
print "=========Supplier============"
outfile = open('supplier_count', 'w')
for supplier in supplier_dic:
    if supplier_dic[supplier] > 1:
        print supplier + ',' + str(supplier_dic[supplier])
        #pass
print "=========Consumer============"
outfile = open('consumer_count', 'w')
for consumer in consumer_dic:
    if consumer_dic[consumer] > 1:
        print consumer + ',' + str(consumer_dic[consumer])
        pass
print "=========Amount===="
print amount_count
print "=====province========"
for prov in prov_dic:
    if prov_dic[prov] > 1:
        print prov + ',' + str(prov_dic[prov])
        pass
```
