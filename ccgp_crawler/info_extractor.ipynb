```python
#encoding = utf-8
import glob
files = glob.glob('./text_data/*')
poss_items = ["采购人", "采购单位", "委托单位"]

for file_name in files:
    info = open(file_name)#2068
    found = False
    for n, line in enumerate(info):
        if n < 2:
            continue
        if ':' in line:
            line = line.strip().split(':')
        else:
            line = line.strip().split('：')
        for item in poss_items:
            if item in line[0]:
                found = True
        if found:
            if len(line) > 1:
                print file_name + '|' + line[1].strip()
            else:
                print file_name + '|' + next(info).strip()
            break
            
    if found == False:
        print file_name + '|' + 'fail'
    
                
    
    
    
```
